## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
# subtotal_client is the same thing as line_subtotal !!!!!!!!!!!!!!!!!!!!!!!!!! :D:D:D:D

client_dataset_df['line_subtotal'] = client_dataset_df['unit_price'] * client_dataset_df['qty']
client_dataset_df[["unit_price", "qty", "line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

client_dataset_df["total_weight"] = client_dataset_df["qty"]*client_dataset_df['unit_weight']
client_dataset_df["shipping_price"] = client_dataset_df["total_weight"].apply(lambda total_weight: 7*total_weight if total_weight>50  else total_weight*10)

client_dataset_df[["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]].head()

```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
client_dataset_df["line_price"] = round((client_dataset_df['line_subtotal'] + client_dataset_df['shipping_price']) * 1.0925, 2)


#client_dataset_df['sales_tax'] = 1.0925
#client_dataset_df['total_price'] = client_dataset_df(["shipping_price"]+client_dataset_df['line_subtotal'])
#lient_dataset_df()
client_dataset_df[["line_subtotal", "shipping_price", "line_price"]].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).


client_dataset_df['line_cost'] = (client_dataset_df['unit_cost'] * client_dataset_df['qty']) + (client_dataset_df['shipping_price']) 

client_dataset_df.head(3)

#i called the df.head(), but just called those three so its easy to read for grader :D !!!! :D 
```

```python
# Create a column for the profit of
# each line using line cost and line price

client_dataset_df['line_profit'] = (client_dataset_df['line_price'] - client_dataset_df['line_cost']) 

client_dataset_df.head(3)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Assuming summary_df is your DataFrame with financial data
# Currency formatting function
def format_currency(value):
    return f"${value/1000000:.2f}M"
#format_millions = lambda x: f'{x/1000000:.2f}M'                    IDK IF I NEED TO


# Format and rename columns
client_summary_data['qty'] = client_summary_data['qty'].apply(lambda x: f'{x:.2f}M')
client_summary_data['shipping_price'] = client_summary_data['shipping_price'].apply(lambda x: f'{x:.2f}M')
client_summary_data['line_price'] = client_summary_data['line_price'].apply(lambda x: f'{x:.2f}M')
client_summary_data['line_profit'] = client_summary_data['line_profit'].apply(lambda x: f'{x:.2f}M') 


# Rename columns
summary_df = client_summary_data.rename(columns={
    'qty': 'Total Units Purchased (Millions)',
    'shipping_price': 'Total Shipping Price (Millions)',
    'line_price': 'Total Revenue (Millions)',
    'line_profit': 'Total Profit (Millions)'
})

# Display the formatted DataFrame
print(summary_df)
```
